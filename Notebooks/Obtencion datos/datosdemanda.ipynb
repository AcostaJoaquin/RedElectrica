```python
import requests

import pandas as pd

from datetime import datetime

from pprint import pprint

from datetime import datetime, timedelta

import plotly.express as px
import plotly.colors as pc
```

```python
lang          = 'es'

start_date1   = '2024-10-29'
end_date1     = '2024-01-01'

start_date2   = '2023-12-31'
end_date2     = '2023-01-01'

start_date3   = '2022-12-31'
end_date3     = '2022-01-01'
```

```python
headers = {
    "Accept": "application/json",
    "Content_Type": "application/json",
    "Host": "apidatos.ree.es"
}
```

```python
def demanda_datos(lang, start_date1, end_date1, start_date2, end_date2, start_date3, end_date3):


    #Demanda 2024
    query = f"start_date={end_date1}T00:00&end_date={start_date1}T23:59&time_trunc=day"

    endpoint = f"https://apidatos.ree.es/{lang}/datos/demanda/evolucion?{query}"
    response = requests.get(url = endpoint, headers = headers)
    data = response.json()

    datetime_lista = list()
    value_lista = list()
    percentage_lista = list()



    for value in data['included']:
        for content in value['attributes']['values']:

                porcentaje = content['percentage']
                valor = content['value']

                fecha = content ['datetime']
                fecha = pd.to_datetime(fecha)
                fecha = fecha.strftime("%d/%m/%Y")

                datetime_lista.append(fecha)
                value_lista.append(valor)
                percentage_lista.append(porcentaje)


    df_demanda1 = pd.DataFrame()
    df_demanda1['Fecha actualización'] = datetime_lista
    df_demanda1['Energia_consumida'] = value_lista


    #Demanda 2023
    query = f"start_date={end_date2}T00:00&end_date={start_date2}T23:59&time_trunc=day"

    endpoint = f"https://apidatos.ree.es/{lang}/datos/demanda/evolucion?{query}"
    response = requests.get(url = endpoint, headers = headers)
    data = response.json()

    datetime_lista = list()
    value_lista = list()
    percentage_lista = list()



    for value in data['included']:
        for content in value['attributes']['values']:

                porcentaje = content['percentage']
                valor = content['value']

                fecha = content ['datetime']
                fecha = pd.to_datetime(fecha)
                fecha = fecha.strftime("%d/%m/%Y")

                datetime_lista.append(fecha)
                value_lista.append(valor)
                percentage_lista.append(porcentaje)


    df_demanda2 = pd.DataFrame()
    df_demanda2['Fecha actualización'] = datetime_lista
    df_demanda2['Energia_consumida'] = value_lista

    #Demanda 2022
    query = f"start_date={end_date3}T00:00&end_date={start_date3}T23:59&time_trunc=day"

    endpoint = f"https://apidatos.ree.es/{lang}/datos/demanda/evolucion?{query}"
    response = requests.get(url = endpoint, headers = headers)
    data = response.json()

    datetime_lista = list()
    value_lista = list()
    percentage_lista = list()



    for value in data['included']:
        for content in value['attributes']['values']:

                porcentaje = content['percentage']
                valor = content['value']

                fecha = content ['datetime']
                fecha = pd.to_datetime(fecha)
                fecha = fecha.strftime("%d/%m/%Y")

                datetime_lista.append(fecha)
                value_lista.append(valor)
                percentage_lista.append(porcentaje)


    df_demanda3 = pd.DataFrame()
    df_demanda3['Fecha actualización'] = datetime_lista
    df_demanda3['Energia_consumida'] = value_lista

    #Concatenación de demandas de diferentes años
    df_demanda = pd.concat([df_demanda3, df_demanda2, df_demanda1], ignore_index = True)

    return df_demanda



df_demanda = demanda_datos(lang, start_date1, end_date1, start_date2, end_date2, start_date3, end_date3)


df_demanda
```

```python
df_demanda.to_csv('demanda_evolucion.csv')
```
